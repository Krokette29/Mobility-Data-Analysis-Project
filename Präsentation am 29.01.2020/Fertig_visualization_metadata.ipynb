```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
```

```
mode = df['mode'].values
num_tracks = df['num_tracks'].values
```

```
# Histrogram
# create figure
fig, ax = plt.subplots(figsize=(15, 10))

# set title
plt.title('num_tracks per mode')

ax.bar(mode, num_tracks, label='number of tracks per mode')

plt.setp( ax.xaxis.get_majorticklabels(), rotation=45 )

for a,b in zip(mode,num_tracks):  
 plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=14) 

ax.legend(loc='upper right')


plt.xlabel('mode')
plt.ylabel('num_tracks')

ax.set_axisbelow(True)

ax.grid(axis='y')

plt.show()
```
