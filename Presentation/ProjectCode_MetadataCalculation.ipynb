## 1.c Average and maximum velocity / acceleration per mode

```python
average_velocity_per_mode = {}

for item in mode_set:
    average_speed = np.mean(gdf.loc[gdf['mode'] == item]['speed'])
    average_velocity_per_mode[item] = average_speed
#     print('average speed of mode {}: {} km/h'.format(item, round(average_speed, 2)))
```

```python
average_acceleration_per_mode = {}

for item in mode_set:
    average_acceleration = np.mean(gdf.loc[gdf['mode'] == item]['acceleration'])
    average_acceleration_per_mode[item] = average_acceleration
#     print('average acceleration of mode {}: {} m/s^2'.format(item, round(average_acceleration, 5)))
```

```python
def calculate_box_plot_characteristics(my_list: list):
    result = {}
    
    result["minimum"] = np.min(my_list)
    result["maximum"] = np.max(my_list)
    result["median"] = np.median(my_list)
    
    q1 = np.percentile(my_list, 25)
    q3 = np.percentile(my_list, 75)
    iqr = q3 - q1
    result["lower_quartile"] = q1
    result["upper_quartile"] = q3
    
    lower_whisker = q1 - 1.5 * iqr
    upper_whisker = q3 + 1.5 * iqr
    rpa_sort = np.sort(my_list)
    for i in range(len(rpa_sort)):
        if rpa_sort[i] > lower_whisker:
            result["lower_whisker"] = rpa_sort[i]
            break
    for i in reversed(range(len(rpa_sort))):
        if rpa_sort[i] < upper_whisker:
            result["upper_whisker"] = rpa_sort[i]
            break
    
    return result
```

```python
max_velocity_per_mode = {}

for item in mode_set:
    max_velocity = calculate_box_plot_characteristics(gdf.loc[gdf['mode'] == item].speed.values)['upper_whisker']
    max_velocity_per_mode[item] = max_velocity
#     print("maximum speed of mode {} is {} km/h".format(item, round(res, 2)))
```

```python
max_acceleration_per_mode = {}

for item in mode_set:
    max_acceleration = calculate_box_plot_characteristics(gdf.loc[gdf['mode'] == item].acceleration.values)['upper_whisker']
    max_acceleration_per_mode[item] = max_acceleration
#     print("maximum acceleration of mode {} is {} m/s^2".format(item, round(res, 5)))
```

```python
def dicts_to_csv(dicts, names, file_name):
    l = [[] for i in range(len(dicts) + 1)]
    
    for item in mode_set:
        l[0].append(item)
        for i in range(len(dicts)):
            l[i + 1].append(dicts[i][item])
    
    df_d = {}
    for i in range(len(names)):
        df_d[names[i]] = l[i]
    df = pd.DataFrame(df_d)
    df.to_csv("./export_metadata/" + file_name, index=False)
    return df
```

```python
df = dicts_to_csv([average_velocity_per_mode, average_acceleration_per_mode, max_velocity_per_mode, max_acceleration_per_mode], 
                  ['mode', 'average_velocity', 'average_acceleration', 'max_velocity', 'max_acceleration'], 'average_and_maximum_velocity_and_acceleration_per_mode.csv')
```
